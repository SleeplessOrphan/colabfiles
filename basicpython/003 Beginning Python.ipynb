### Step 4: Performing the Calculation

Now we need to perform the calculation based on the operation the user has selected. We use if and elif statements to decide what operation to perform.

```python
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 == 0:
        print("Cannot divide by zero")
        exit()
    result = num1 / num2
else:
    print("Invalid operation")
    exit()
```
The `if num2 == 0:` line checks if the user is trying to divide by zero, which is not allowed in mathematics. If so, we display an error message and stop the program with `exit()`.
