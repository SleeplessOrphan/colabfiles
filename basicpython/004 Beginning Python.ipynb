### Part 1: Understanding the Problem

Before writing any code, it's important to understand what we are trying to solve. We want a program that can convert temperatures between Celsius, Fahrenheit, and Kelvin. To do this, you need to know the mathematical formulas for conversion between these units.

**Celsius to Fahrenheit:**

When converting from Celsius to Fahrenheit, the formula is
$$ F = \frac{9}{5}C + 32 $$

**Fahrenheit to Celsius:**

When converting from Fahrenheit to Celsius, the formula is
$$ C = \frac{5}{9}(F - 32) $$

**Celsius to Kelvin:**

When converting from Celsius to Kelvin, the formula is
$$ K = C + 273.15 $$

**Kelvin to Celsius:**

When converting from Kelvin to Celsius, the formula is
$$ C = K - 273.15 $$

# Step 3: Implement Conversion Logic

Here, we'll use if-elif-else statements to decide which formula to use based on the unit provided.

```python
if unit == 'C':
    fahrenheit = (temp * 9/5) + 32
    kelvin = temp + 273.15
    print(f"The temperature in Fahrenheit is {fahrenheit} and in Kelvin is {kelvin}.")

elif unit == 'F':
    celsius = (temp - 32) * 5/9
    kelvin = celsius + 273.15
    print(f"The temperature in Celsius is {celsius} and in Kelvin is {kelvin}.")

elif unit == 'K':
    celsius = temp - 273.15
    fahrenheit = (celsius * 9/5) + 32
    print(f"The temperature in Celsius is {celsius} and in Fahrenheit is {fahrenheit}.")

else:
    print("Invalid unit. Please enter C, F, or K.")
```
